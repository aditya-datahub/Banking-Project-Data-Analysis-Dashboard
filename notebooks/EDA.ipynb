```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("Banking_Clients.csv")
df.head()
```

```python
df.info()
```

```python
# Generating descriptive statistics for the dataframe
df.describe()
```

```python
df['Estimated Income'].min()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Brand'] = pd.cut(df['Estimated Income'], bins = bins, labels= labels, right= False)

df['Income Brand'].value_counts().sort_index().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical columns

categorical_cols = df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", 
                       "Properties Owned", "Risk Weighting", "Income Brand"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    print(df[col].value_counts())
    print("-" * 45)
```

```python
# Uniavriate Analysis

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure",
                                  "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Brand"]]):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
    plt.show()  
```

```python
# Bivariate Analysis

for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", 
                                  "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Brand"]]):
    
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='Nationality')
    plt.show() 
```

```python
# Histplot of value counts for different Columns

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize = (6,4))
    sns.histplot(df[col])
    plt.title(f"Histogram of {col} Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings']

# Ultravariate analysis and visualization
plt.figure(figsize = (10,10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2,1,i+1)
    sns.histplot(df[col], kde = True)
    plt.title(col)
plt.show()
```

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
                  'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10,10))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
'''
Insights of EDA:

The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts", and "Foreign Currency Account" 
indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.    
'''
```
